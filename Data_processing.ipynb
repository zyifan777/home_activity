```python
import matplotlib.pyplot as plt
import numpy as np
import statistics as stats
```

```python
getOnBed = open('data/getOnBed.csv','r')
getOutBed = open('data/getOutBed.csv','r')
layOnBed = open('data/layOnBed.csv','r')
sitOnBed = open('data/sitOnBed.csv','r')
```

```python
d1 = [int(x) for x in getOnBed.readline().split(',')]
d2 = [int(x) for x in getOutBed.readline().split(',')]
d3 = [int(x) for x in layOnBed.readline().split(',')]
d4 = [int(x) for x in sitOnBed.readline().split(',')]
```

```python
print len(d1),len(d2),len(d3),len(d4)
```

```python
plt.figure(figsize=(20,12))
plt.subplot(2,2,1)
plt.title("Get On Bed")
plt.xlabel("time(ms)")
plt.plot(d1)

plt.subplot(2,2,2)
plt.title("Get out of Bed")
plt.xlabel("time(ms)")
plt.plot(d2)

plt.subplot(2,2,3)
plt.title("Lay on Bed")
plt.xlabel("time(ms)")
plt.plot(d3)

plt.subplot(2,2,4)
plt.title("Sit on Bed")
plt.xlabel("time(ms)")
plt.plot(d4)

plt.show()
```

## Simple Feature Extraction

```python
print "Mean: ",stats.mean(d1),stats.mean(d2),stats.mean(d3),stats.mean(d4)
print "Median: ", stats.median(d1),stats.median(d2),stats.median(d3),stats.median(d4)
print "Variance: ", stats.variance(d1),stats.variance(d2),stats.variance(d3),stats.variance(d4)
print "Min: ", min(d1),min(d2),min(d3),min(d4)
print "Max: ", max(d1),max(d2),max(d3),max(d4)
```

## Simple Feature Extraction within Phases

```python
label = ["Get On Bed","Get Out of Bed", "Lay on Bed", "Sit on Bed"]
for d,l in zip([d1,d2,d3,d4],label):
    print l
    for i in range(3):
        i = i*20000;
        print str.format("Mean:{:5.5}  Median:{:5.7}  Var:{}  Min:{:5}  Max:{:5}",stats.mean(d[i:i+20000]), 
                        stats.median(d[i:i+20000]), stats.variance(d[i:i+20000]), 
                        min(d[i:i+20000]), max(d[i:i+20000]))
    print '\n'
    
```
